Demonstrar a execução do cálculo e utilizar um comentário para afirmar que é um programa de exemplo de folha de pagamento.

```java
import java.util.Scanner;
import java.text.DecimalFormat; 

Scanner sc = new Scanner(System.in);

System.out.print("Digite o nome do funcionário: ");
String nome = sc.nextLine();

System.out.print("Digite o número de horas trabalhadas: ");
int horasTrabalhadas = sc.nextInt();

System.out.print("Digite o valor da hora de trabalho: ");
double valorHora = sc.nextDouble();

double salarioBruto = horasTrabalhadas * valorHora;
double inss = salarioBruto * 0.11;
double ir = salarioBruto * 0.15;
double salarioLiquido = salarioBruto - inss - ir;

DecimalFormat dFormat = new DecimalFormat("0.00");// transforma um inteiro em decimal

System.out.println("\nFolha de pagamento:");
System.out.println("Funcionário: " + nome);
System.out.println("Salário bruto: R$" + dFormat.format(salarioBruto));
System.out.println("INSS: R$" + dFormat.format(inss));
System.out.println("IR: R$" + dFormat.format(ir));
System.out.println("Salário líquido: R$" + dFormat.format(salarioLiquido));
    

```
